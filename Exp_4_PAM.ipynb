```python
import matplotlib.pylab as plt
import matplotlib.pyplot as plot
import numpy as num
from scipy import signal
```

```python
a = int(input('Enter the amplitude of modulating signal= '))
f = int(input('Enter the frequency of modulating signal  = '))
```

Generating train of impulses, sine wave and PAM wave

```python
t = num.arange(0,2,0.02)  # for a total of 16 samples
t_1=t                      

# %generation of an impulse signal, modulating signal and PAM signal
x1=[]
for i in range(len(t)):
    x1.append(1)                              #X1 IS THE TRAIN OF IMPULSES WHICH IS CARRIER
x2 = a*num.sin(2*num.pi*f*t)                  # X2 IS THE MODULATING SIGNAL WHICH IS SINE WAVE
y = x1*x2;                                    #PAM SIGNAL ideally sampled
```

```python
#for impulse signal plot
plt.stem(t,x1);
plt.title('Impulse Signal');
plt.xlabel('Time');
plt.ylabel('Amplitude ');
plt.grid(True)
plt.show()
```

```python
#for modulating signal or sine wave plot
plt.plot(t,x2); 
plt.title('Modulating signal or Sine Wave');
plt.xlabel('Time ');
plt.ylabel('Amplitude ');
plt.grid(True)
plt.show()
```

```python
#for PAM signal plot
plt.stem(t,y);
plt.title('PAM Signal');
plt.xlabel('Time');
plt.ylabel('Amplitude ');
plt.grid(True)
plt.show()
```

```python
plt.subplots_adjust(bottom=-2)
plt.subplot(3,1,1)
plt.stem(t, x1,'r')
plt.ylabel('amplitude')
plt.xlabel('t')
plt.title('Impulse signal')

plt.subplot(3,1,2)
plt.plot(x2, 'b')
plt.ylabel('amplitude')
plt.xlabel('t')
plt.title('mod')

plt.subplot(3,1,3)
plt.stem(t, y, 'purple')
plt.ylabel('amplitude')
plt.xlabel('t')
plt.title('PAM signal')
```

NATURAL PAM

```python
square = signal.square(80 * t)
square[square < 0] = 0                      # All negative values in the AM signal are made zero which means it is diode rectification

plt.plot(t, square,'r')
```

```python
y2 = square*x2
#for Natural PAM wave plot
plt.plot(t,y2);
plt.title('Natural PAM Wave');
plt.xlabel('Time');
plt.ylabel('Amplitude');
plt.grid(True)
plt.show()
```

```python
plt.subplots_adjust(bottom=-2)
plt.subplot(3,1,1)
plt.stem(t, x1,'g')
plt.ylabel('amplitue')
plt.xlabel('t')
plt.title('Sampling pulses')

plt.subplot(3,1,2)
plt.plot(t,x2, 'r')
plt.ylabel('amplitude')
plt.xlabel('t')
plt.title('modulating signal')

plt.subplot(3,1,3)
plt.plot(t, y2, 'purple')
plt.ylabel('amplitude')
plt.xlabel('t')
plt.title(' Natural PAM signal')
```
